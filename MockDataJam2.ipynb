### Making Predictions With Our K Nearest Neighbors Algorithm

```python
import pandas #fixed
from sklearn.impute import KNNImputer

dfr.replace('', pd.NA, inplace=True)

imputer = KNNImputer(n_neighbors=3)

df_full = pd.DataFrame(imputer.fit_transform(dfr), columns=dfr.columns)

display(dfr)
```

```python
display(df_full)
```

# Regression Analysis : 

```python
print(df_full.columns)
```

```python
from sklearn.linear_model import LinearRegression

x = df_full[["GII(Gender Inequality Index)" , "Maternal_mortality" , "Adolescent_birth_rate" , "Seats_parliament" , "F_secondary_educ" , "M_secondary_educ" , "F_Labour_force",
             "M_Labour_force" , "Population"]]
y = df_full["GDP   per capita"]

model = LinearRegression()

model.fit(x,y)

coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

```python
import statsmodels.api as sm

X = sm.add_constant(df_full[["GII(Gender Inequality Index)", "Maternal_mortality", "Adolescent_birth_rate", "Seats_parliament", "F_secondary_educ", 
                             "M_secondary_educ", "F_Labour_force", "M_Labour_force", "Population"]])
y = df_full["GDP   per capita"]

model = sm.OLS(y, X)

results = model.fit()

print(results.summary())
```
