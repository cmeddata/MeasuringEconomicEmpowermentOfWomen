## Description of Variables

### Dependent Variable 
1. **GDP Per Capita**: Gross Domestic Product (GDP) per capita (per person) is the total GDP divided by the population. It is an important indicator of economic performance and a useful unit to make cross-country comparisons of average living standards and economic well being. 


### Independent Variables
1. **Country**: This is the name of the country being studied. It serves as the primary identifier for each data point in our analysis.

2. **Gender Inequality Index**: This variable measures the extent of gender inequality in a country, encompassing factors like women's economic participation, education, health, and political empowerment. It's typically expressed as a numerical score, with higher values indicating greater gender inequality.

3. **Seats in Parliament**: This variable represents the number of parliamentary seats held by women in the country's legislature. It reflects the level of women's political representation.

4. **Female Secondary Education**: This variable quantifies the level of secondary education attained by females in the country, typically measured as a percentage of the female population.

5. **Male Secondary Education**: Similar to the previous variable, this one represents the level of secondary education attained by males in the country, often expressed as a percentage.

6. **Male Labour Force**: This variable indicates the percentage of the male population that is engaged in the labor force, providing insights into the male workforce's participation.

711. **Female Labour Force**: This variable represents the percentage of the female population actively participating in the labor force, including those employed or seeking employment. It provides crucial insights into the level of women's economic engagement within the country and is a significant aspect of gender inequality and economic empowerment analysis.

7. **Land Area**: Land area represents the total geographic area of the country in square kilometers. It's a fundamental geographic variable.

8. **Median Age**: Median age is the age at which exactly half of the population is older, and half is younger. It provides information about the age distribution of the population.

9. **Urban Population**: This variable signifies the percentage of the country's population residing in urban areas, giving insight into the level of urbanization.

10. **Maternal Mortality Rate**: The maternal mortality rate is the number of maternal deaths per 100,000 live births in a given year. It's an important health indicator that reflects the safety of maternal health.

## Reading the csv file using python pandas library

```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("EconomicEmpowermentOfWomen.csv")
```

```python
dfr = pd.read_csv("RawDataEcon.csv")
```

### Making Predictions With Our K Nearest Neighbors Algorithm

```python
from sklearn.impute import KNNImputer

dfr.replace('', pd.NA, inplace=True)

imputer = KNNImputer(n_neighbors=3)

df_full = pd.DataFrame(imputer.fit_transform(dfr), columns=dfr.columns)

display(dfr)
```

```python
display(df_full)
```

# Regression Analysis : 

```python
print(df_full.columns)
```

```python
from sklearn.linear_model import LinearRegression

x = df_full[['GII(Gender Inequality Index)', 'Rank', 'Maternal_mortality',
       'Adolescent_birth_rate', 'Seats_parliament', 'F_secondary_educ',
       'M_secondary_educ', 'F_Labour_force', 'M_Labour_force', 'Population',
       'Yearly Change', 'Net Change', 'Density', 'Land Area ',
       'Migrants  (net)', 'Fert.  Rate', 'Med.  Age', 'Urban  Pop %',
       'World  Share', 'GDP  (nominal, 2022)', 'GDP   growth',
       'Population  (2022)', 'GDP   per capita', 'Share of  World GDP']]


y = df_full["GDP   per capita"]

model = LinearRegression()

model.fit(x,y)

coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

```python
import statsmodels.api as sm

X = sm.add_constant(df_full[['GII(Gender Inequality Index)', 'Maternal_mortality',
       'Adolescent_birth_rate', 'Seats_parliament', 'F_secondary_educ',
       'M_secondary_educ', 'F_Labour_force', 'M_Labour_force', 'Population',
       'Yearly Change', 'Net Change', 'Density', 'Land Area ',
       'Migrants  (net)', 'Fert.  Rate', 'Med.  Age', 'Urban  Pop %']])

y = df_full['GDP   per capita']

model_0 = sm.OLS(y, X)

results = model_0.fit()

print(results.summary())
```

```python
import statsmodels.api as sm

X = sm.add_constant(df_full[['GII(Gender Inequality Index)','Seats_parliament','F_secondary_educ','M_secondary_educ','F_Labour_force','M_Labour_force','Density','Land Area ','Med.  Age','Urban  Pop %','Maternal_mortality']])


y = df_full['GDP   per capita']

model = sm.OLS(y, X)

results = model.fit()

print(results.summary())
```
